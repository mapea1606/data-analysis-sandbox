# Reindexing

```python
obj
```

```python
obj2 = obj.reindex(["a", "b", "c", "d", "e"])
```

```python
obj2
```

```python
obj3 = pd.Series(["blue", "purple", "yellow"], index=[0, 2, 4])
```

```python
obj3.reindex(np.arange(6), method="ffill")
```
